# **RaspberryPi aufsetzten für Projektarbeiten an der NTB**

Erstellt am 7.5.2018 für Raspberry Pi 3 Model B.

Mit SDformatter (geht auch für Windows10) die SD-Karte formatieren.

Von [www.raspberrypi.org](http://www.raspberrypi.org) aktuelles NOOBS. „Network install only“ (nicht offline). Voraussetzung ist Netzwerk-Zugriff. Anschliessend auf SD-Karte kopieren/entpacken.

Im Klassenverbund das WLAN «gast-ntb» verwenden. Das tagesgültige Passwort ist im Intranet.

Dann SD-Karte in RaspberryPi einsetzen und starten. Zuerst Sprache Deutsch und Tastatur CH eingeben, dann WLAN verbinden (hierzu WLAN user/pwd eingeben, wofür die richtige Tastatur gewählt werden muss um ggf Sonderzeichen eingeben zu können). Raspbian with PIXEL (Desktop Version) ankreuzen, und „Installieren“ klicken. Installation dauert über 1 Stunde!

Nach Installation erfolgt ein Neustart und es ist ein Login erforderlich. Prüfe ob die richtige Tastatur konfiguriert wurde mit Eingabe „z“ und „y“.
Login: pi; Password: raspberry.

Nun konfigurieren wir den RaspberryPi über das raspi-config Tool. Hierzu kann die Console geöffnet werden:

    sudo raspi-config

Das Konfigurationstool kann auch über den Desktop geöffnet über „Einstellungen > Raspberry-Pi-Konfiguration"

Folgende Einstellungen vornehmen:
* SSH aktivieren in  „Interface Option“ (oder "Advanced Options") unter "SSH" enable, somit können wir später remote über ssh zugreifen.
* Hier im Menü „Boot Option“ die Desktop-Option und autologin wählen (default). 
* Tastatur und Sprache kann ebenfalls hier konfiguriert werden eingeben.

Neues Passwort: um nicht jedes Mal das lange Passwort „raspberry“ mit der „y“ oder „z“ Unsicherheit eingeben zu müssen, ändern wir das Passwort zu „??passwort wählen??“ (Achtung Kleinschreibung beachten). Änderung über die Konsole. Für den user „pi“ (Standard):

    passwd

Und für den root-user:

    sudo passwd


Dann wird das System aktualisiert. Dieser Schritt empfiehlt sich allgemein bei OpenSource-SW; immer auf dem aktuellem Stand arbeiten. Zuerst die „system’s package list“ updaten und Spyder installieren. Hierzu werden die Befehle in der Console eingegeben.

    sudo apt-get update

    sudo apt-get install spyder3

## Serielle Schnittstelle für GPIO
Für die Aktivierung der seriellen Schnittstelle auf den GPIO-Pins (14, 15) muss Bluetooth deaktivert werden, da Bluetooth diese Position standardmässig belegt. Hierfür ergänzen wir ein Befehl in der config.txt Datei. Als Editior verwenden wir „nano“ aufgerufen über die Console:

    sudo nano /boot/config.txt

Am Ende der Datei ergänzen wir folgende zwei Befehlw: 

    dtoverlay=pi3-disable-bt

    enable_uart=1

Speichern und Schliessen.

Anschliessend wird die serielle Console deaktiviert über Einstellungen in der Datei „cmdline.txt“.

    sudo nano /boot/cmdline.txt

Wir entfernen die Einträge welche wie folgt aussehen: `console=serial0,115200` oder `console=ttyAMA0,115200`
Nach Speicher/Schliessen, starten wir das System neu:

    sudo reboot

Nun müsste das Bluetooth-Symbol rechts oben deaktiv sein.

## File Transfer 
Um bestehende py-Skripts vom PC auf den RaspberryPi zu übertragen benötigen wir einen Remote-Zugriff. 
Hierzu verwenden wir das auf Windows installierte Tool «WinSCP».
Windowstaste drücken und «WinSCP» eingeben. Port: 22, Username: pi, Password: ??passwort wählen??.

Die IP-Adresse des RPI erhält man durch Eingabe auf dem RPI:

    ifconfig

Zur Info: MAC-Adresse abfragen

    netstat -ei

Alternativen zu WinSCP sind «Bitvise SSH Client» oder "FileZilla".

## Remote Desktop
Für den Remote Desktop Zugriff müssen erst folgende Pakete deinstalliert werden, sodass eine definierte Ausgangslage ist (im Normalfall sind einige angeführten Pakete nicht installiert):

    sudo apt-get remove xrdp vnc4server tightvncserver

anschliessend werden die Pakete in richtiger Reihenfolge installiert. Zuerst:

    sudo apt-get install tightvncserver

dann:

    sudo apt-get install xrdp

Beim Remote Desktop Zugriff wird die englische Tastatur (km-0409.ini) geladen, wir wollen aber die schweizer Tastatur (km-0807.ini). Hierzu starten wird den Editor, allerdings über die Console sodass wir „root“-Rechte (sudo) haben:

    sudo pcmanfm

Darstellung bei linker Spalte von „Ort“ auf „Verzeichnisbaum“ stellen. Navigieren in den Ordner `/etc/xrdp`.
Dort nimmt er standardmässig die Datei „km-0409.ini“. Wir suchen im Internet die Datei „km-0807.ini“. Wir öffnen nochmals ein Fenster: sudo pcmanfm
Und kopieren unter „Downloads“ die Datei in den Ordner `/etc/xrdp`.
Dann umbenennen „km-0409.ini“  in „km-0409_orginal.ini“ 
und umbenennen „km-0807.ini“  in „km-0409.ini“.

### Remote Desktop Zugriff
Auf dem PC kann nun Remote Desktop-Connection gestartet werden. Es öffnet sich ein Fenster, bei welchem das Modul "sesman.Xvnc" gewählt wird. Der Login lautet:
Username: pi
Password: ??passwort wählen?? (Achtung prüfen ob englische Tastatur, d.h. z-Taster für y)
 

Nun können die installierten Pakete upgraden werden. Im Klassenverbund nicht eingeben, da es zu lange dauert.

    sudo apt-get dist-upgrade

    sudo reboot
